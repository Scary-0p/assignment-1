```python
# 26. Twin primes between 1 and 100

primes = []
for num in range(2, 101):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

for i in range(len(primes)-1):
    if primes[i+1] - primes[i] == 2:
        print((primes[i], primes[i+1]))
```

```python
# 27. Harshad (Niven) number check
num = int(input("Enter a number: "))
digit_sum = 0
for ch in str(abs(num)):
    digit_sum += int(ch)
if digit_sum != 0 and num % digit_sum == 0:
    print(num, "is a Harshad number")
else:
    print(num, "is NOT a Harshad number")
```

```python
# 28. Pascal’s Triangle up to n rows
n = int(input("Enter number of rows for Pascal's Triangle: "))
for r in range(n):
    
    val = 1
    row = []
    for k in range(r + 1):
        row.append(str(val))
        
        val = val * (r - k) // (k + 1)
    print(' '.join(row).center(4*n))
```

```python
# 30. Strong number check (sum of factorials of digits equals number)
import math
num = int(input("Enter a number: "))
s = 0
for ch in str(num):
    d = int(ch)
    s += math.factorial(d)
if s == num:
    print(num, "is a Strong number")
else:
    print(num, "is NOT a Strong number")
```

```python
# 31. Reverse a number using while, then check if reversed number is prime
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

num = int(input("Enter a number: "))
n = abs(num)
rev = 0
while n > 0:
    rev = rev * 10 + (n % 10)
    n //= 10

print("Reversed number:", rev)
print("Reversed is prime?" , is_prime(rev))
```

```python
# 33. Check Duck number using a while loop (contains zero but not starting with zero)
s = input("Enter the number (preserve leading zeros if any): ").strip()
if s == "":
    print("No input")
else:
    if s[0] == '0':
        print("Not a Duck number (starts with 0).")
    else:
        i = 0
        has_zero = False
        while i < len(s):
            if s[i] == '0':
                has_zero = True
                break
            i += 1
        print("Duck number?" , has_zero)
```

```python
# 34. Check if a number is a Happy number (while loop)
def next_sum_sq_digits(n):
    s = 0
    while n > 0:
        d = n % 10
        s += d*d
        n //= 10
    return s

num = int(input("Enter a number: "))
seen = set()
n = num
while n != 1 and n not in seen:
    seen.add(n)
    n = next_sum_sq_digits(n)

if n == 1:
    print(num, "is a Happy number")
else:
    print(num, "is NOT a Happy number")
```

```python
# 35. Largest prime factor using while loop
n = int(input("Enter a number (>1): "))
num = n
largest = -1


while num % 2 == 0:
    largest = 2
    num //= 2


i = 3
while i * i <= num:
    while num % i == 0:
        largest = i
        num //= i
    i += 2

if num > 2:
    largest = num

print("Largest prime factor of", n, "is", largest)
```

```python
# 36. Keep accepting strings until the entered string is a palindrome (using while)
while True:
    s = input("Enter a string: ")
    i = 0
    j = len(s) - 1
    is_pal = True
    while i < j:
        if s[i] != s[j]:
            is_pal = False
            break
        i += 1
        j -= 1
    if is_pal:
        print("You entered a palindrome:", s)
        break
    else:
        print("Not a palindrome, try again.")
```

```python
# 37. Compute digital root (sum digits until single-digit) using while
n = int(input("Enter a number: "))
n = abs(n)
while n >= 10:
    s = 0
    temp = n
    while temp > 0:
        s += temp % 10
        temp //= 10
    n = s
print("Digital root:", n)
```

```python
# 38. Generate the Collatz sequence using while (until n = 1)
n = int(input("Enter starting positive integer: "))
if n <= 0:
    print("Enter a positive integer.")
else:
    seq = []
    while n != 1:
        seq.append(n)
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
    seq.append(1)
    print("Collatz sequence:", seq)
```

```python
# 39. Check Kaprekar number using while loop to examine splits
num = int(input("Enter a number: "))
sq_str = str(num * num)
i = 1
is_kaprekar = False

while i < len(sq_str):
    left = int(sq_str[:i]) if sq_str[:i] != "" else 0
    right = int(sq_str[i:]) if sq_str[i:] != "" else 0
    if left + right == num:
        is_kaprekar = True
        break
    i += 1

if num == 1:
    is_kaprekar = True

print(num, "is Kaprekar?" , is_kaprekar)
```

```python
20) Simple ATM simulation using a while loop (Check balance / Deposit / Withdraw / Exit)
balance = 1000.0   

while True:
    print("\n--- ATM Menu ---")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")
    choice = input("Choose an option (1-4): ").strip()

    if choice == '1':
        print(f"Current balance: ₹{balance:.2f}")
    elif choice == '2':
        amt = float(input("Enter amount to deposit: ₹"))
        if amt > 0:
            balance += amt
            print(f"Deposited ₹{amt:.2f}. New balance: ₹{balance:.2f}")
        else:
            print("Invalid amount.")
    elif choice == '3':
        amt = float(input("Enter amount to withdraw: ₹"))
        if amt <= 0:
            print("Invalid amount.")
        elif amt > balance:
            print("Insufficient balance.")
        else:
            balance -= amt
            print(f"Withdrew ₹{amt:.2f}. New balance: ₹{balance:.2f}")
    elif choice == '4':
        print("Thanks for using the ATM. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1-4.")
```
